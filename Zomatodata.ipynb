```python
ratings = final_df.groupby( ['Aggregate rating', 'Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings.head()
```

```python
ratings.groupby(['Aggregate rating','Rating text']).size()
```

# Observation:
    1.Rating 4.9 to 4.5 indicates Excellent.
    2.Rating 4.4 to 4.0 indicates very good.
    3.Rating 3.9 to 3.5 indicates good.
    4.Rating 3.4 to 2.5 indicates Average.
    5.Rating 2.4 to 1.8 indicates poor.

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(20,10)
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color', data=ratings, palette=['purple','red','orange','yellow','green','green'])
```

Observation:
 1. not rated is very high.
 2. max rating is given between 2.9 to 3.9

```python
sns.countplot(x='Rating color',data=ratings,palette=['purple','red','orange','yellow','green','green'])
```
